```python
# Data Visualization

import matplotlib.pyplot as plt

plt.plot(1, 2, 'ko')
plt.plot([0, 2], [0, 4], 'r--')
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.title("The graph title")

_, axs = plt.subplots(1, 2, figsize=(8, 5))
axs[0].plot(np.random.randn(10, 5))
axs[1].imshow(np.random.randn(10, 5))
```

```python
# Translating formulas to code

x = np.arange(-4, 5)
y = x ** 2
plt.plot(x, y, 's-')
plt.show()

x = np.linspace(-4, 4, 42)
y = x ** 2
plt.plot(x, y, 's-')
plt.show()

x = np.linspace(-4, 4, 56)
alpha = 2.73
beta = 4
numerator = alpha
denomiator = 1 + np.exp(-beta * x)
fx = numerator / denomiator
plt.plot(x, fx, 's-')
plt.show()
```

```python
# Print formatting and f-strings

var1 = 10.54
print(f'The variable value is {var1}, which makes me happy.')

the_list = ["Jacob", 7]
print(f'{the_list[0]} eats {the_list[1] * 100} grams of chocolate each day.')

pi = 22/7
print(f'{pi}, {pi:.3f}')
```

```python
# Comparators

print(4 < 5)
print(4 > 5)
print(4 == 5)
```

```python
# If Statements

var = 4
if var == 4:
    print(f"{var} equals 4!")

print("I'm outside the if statement.")

var = 4
if var == 5:
    print(f"{var} equals 5!")

print("I'm outside the if statement.")
```

```python
# Elif and Else Statements

var = 4

if var == 5:
    print("var is 5")
elif var > 5:
    print("var > 5")
else:
    print("var < 5")

print("Outside the if-elif-else")
```

```python
# Multiple Conditions

if 4 == 4 and 4 < 10:
    print("Code example 1.")

if 4 == 5 and 4 < 10:
    print("Code example 2.")

if 4 == 5 or 4 < 10:
    print("Code example 3.")
```

```python
# For Loops

print(1)
print(2)
print(3)

for i in range(0, 10):
    print(i + 1)

the_list = [2, "hello", np.linspace(0, 1, 14)]
for item in the_list:
    print(item)
```

```python
# Nested Control Statements

powers = [0] * 10

for i in range(len(powers)):
    if i % 2 == 0 and i > 0:
        print(f'{i} is an even number')

    if i > 4:
        powers[i] == i ** 2

print(powers)
```

```python
# Measuring Computation Time

import time

clock_start = time.time()
for i in range(0, 10000):
    random_matrix = np.random.randn(100, 100)
    np.linalg.inv(random_matrix)
computation_time = time.time() - clock_start
print(computation_time)
```
